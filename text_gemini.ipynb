```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
GOOGLE_API_KEY=os.getenv("GOOGLE_API_KEY")
```

```python
llm=ChatGoogleGenerativeAI(
    model='gemini-2.5-flash',
    temperature=0.3
)
```

```python
llm.invoke("How Python is better than C++ and why python is slow?").content
```

```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
embed = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
loader=TextLoader("intro.txt")
document=loader.load()
```

```python
document[0].page_content
```

```python
text_splitter=CharacterTextSplitter(chunk_size=30,chunk_overlap=10)
```

```python
chunks=text_splitter.split_documents(document)
```

```python
for c in chunks:
    print(c.page_content)
```

```python
vectorstores=Chroma.from_documents(chunks,embed)
```

```python
retriever=vectorstores.as_retriever(search_type='similarity',search_kwargs={'k':3})
```

```python
from langchain.chains import RetrievalQA
from langchain.memory import ConversationBufferMemory
from langchain.prompts import PromptTemplate
memory=ConversationBufferMemory()
```

```python
template="""You are Text Bot. You wil get a context and based on it you have to reply in a good way.
Here is Input: {question}
Here is Context: {context}"""
prompt=PromptTemplate(template=template,input_variables=['question','context'])
```

```python
text_bot=RetrievalQA.from_chain_type(llm=llm,retriever=retriever,memory=memory,chain_type_kwargs={'prompt':prompt})
```

```python
response=text_bot.invoke("What this about?")
```

```python
print('-------TEXT BOT-------')
print(response['result'])
```
