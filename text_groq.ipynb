```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
GROQ_API_KEY=os.getenv("GROQ_API_KEY")
```

```python
llm=ChatGroq(
    model='deepseek-r1-distill-llama-70b',
    temperature=0.3
)
```

```python
llm.invoke('What is DEEP SEEK').content
```

```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
embed=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
text_loader=TextLoader("intro.txt")
```

```python
document=text_loader.load()
```

```python
document
```

```python
document[0].page_content
```

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=30,chunk_overlap=5)
```

```python
chunks=text_splitter.split_documents(document)
```

```python
vectostores=FAISS.from_documents(chunks,embed)
```

```python
retriever=vectostores.as_retriever(search_type="similarity",search_kwargs={'k':3})
```

```python
from langchain.chains import RetrievalQA
from langchain.memory import ConversationBufferMemory
from langchain.prompts import PromptTemplate
memory=ConversationBufferMemory()
```

```python
template="""You are a text you will get a context and reply based on it.
Here is input: {question}
Here is context: {context}"""
prompt=PromptTemplate(template=template,input_variables=['question','context'])
```

```python
text_bot=RetrievalQA.from_chain_type(llm=llm,memory=memory,retriever=retriever,chain_type_kwargs={'prompt':prompt})
```

```python
response=text_bot.invoke("Who is Muhammad hamza")
```
